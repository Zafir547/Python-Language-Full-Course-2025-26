Dictionaries are used to store data values in key:value pairs

They are unordered, mutable(changeable) & don’t allow duplicate keys

“key” : value

dict[”name”], dict[”cgpa”], dict[”marks”]

dict[”key”] = “value”   #to assign or add new

### **Set in Python**

Set is the collection of the unordered items.

Each element in the set must be unique & immutable.

nums = { 1, 2, 3, 4 }

set2 = { 1, 2, 2, 2 }

#repeated elements stored only onnce, so it resolved to {1, 2}

null_set = set( ) #empty set syntax

```python
collection = {1, 2, 3, 4}

print(collection)
print(type(collection))
```

```python
collection = {1, 2, 2, 2, "hello", "world", 4}

print(collection)
print(type(collection))
```

If you want to sort only numbers or only strings, filter them first:

Sort only numbers:

```python
collection = {1, 2, 2, 2, "hello", "world", 4}
numbers_only = sorted([x for x in collection if isinstance(x, (int, float))])
print(numbers_only)  # Output: [1, 2, 4]
```

Sort only strings:


```python
collection = {1, 2, 2, 2, "hello", "world", 4}
numbers_only = sorted([x for x in collection if isinstance(x, str)])
print(numbers_only)  # Output: ['hello', 'world']
```

```python
collection = {1, 2, 2, 2, "hello", "world", 4}

print(collection)
print(len(collection))
```

```python
collection = {} #empty dictionary

print(type(collection))
```

```python
collection = set() #empty set; syntax

print(type(collection))
```

### **Set Methods**

set.add( el ) #adds an element

```python
collection = set()
collection.add(1)
collection.add(2)
collection.add(2)

print(collection)
```

set.remove( el ) #removes the elem an

```python
collection.remove(1)
print(collection)
```

Len number

```python
collection = set()
collection.add(1)
collection.add(2)
collection.add("apnacollege")
collection.add((1,2,3))
collection.add((1,2,3))

print(collection)
print(len(collection))
```

set.clear( ) #empties the set

```python
collection = set()
collection.add(1)
collection.add(2)
collection.add("apnacollege")
collection.add((1,2,3))
collection.add((1,2,3))

collection.clear()
print(len(collection))
```

set.pop( ) #removes a random value

```python
collection = {"hello", "apnacollege", "world", "coding", "python"}

print(collection.pop())
```

set.union( set2 ) #combines both set values & returns new

```python
set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1.union(set2)) #{1, 2, 3, 4}
```

set.intersection( set2 ) #combines common values & returns new

```python
set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1.intersection(set2)) #{2, 3}
```
