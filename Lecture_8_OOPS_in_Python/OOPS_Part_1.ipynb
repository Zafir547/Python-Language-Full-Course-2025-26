### **_ _init_ _ Function**
 
All classes have a function called _ _init_ _(), which is always executed when the object is being
initiated.

- creating class

    ```bash
    class Student:
        def __init__( self, fullname ):
            self.name = fullname
    ```

- creating object

    ```bash
    s1 = Student("karan")
    print(s1.name)
    ```

*The self parameter is a reference to the current instance of the class, and is used to access variables that belongs to the class.

````python
class Student:
    name = "arham"
    def __init__(self):
        print("Adding new student in database..")

s1 = Student()
# print(s1.name)
````

````python
class Student:
    def __init__(self,fullname): # create constructor (def __init__)
        self.name = fullname
        print("Adding new student in database..")

s1 = Student("arham")
print(s1.name)

s2 = Student("abdul")
print(s2.name)
````

attributes -> data ; variables

````python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
        print("Adding new student in database..")

s1 = Student("arham",97)
print(s1.name,s1.marks)

s2 = Student("abdul",88)
print(s2.name,s2.marks)
````

Default Constructor

```bash
def __init__(self):
    pass
```        

Parameterized Constructor

```bash
def __init__(self,car,model):
    self.car = car
    self.model = model
```

````python
class Student:
   
   #default constructors
    def __init__(self):
        pass
    
    #parameterized constructor
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
        print("Adding new student in database..")

s1 = Student("arham",97)
print(s1.name,s1.marks)

s2 = Student("abdul",88)
print(s2.name,s2.marks)
````
