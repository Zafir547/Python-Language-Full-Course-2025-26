### **Static Methods**

Methods that don’t use the self parameter (work at class level)

```bash
class Student:
    @staticmethod  #decorator
    def college( ):
        print( “ABC College” )
```

*Decorators allow us to wrap another function in order to
extend the behaviour of the wrapped function, without
permanently modifying it


````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Static Methods
    @staticmethod
    def hello():
        print("hello")    

    def get_avg(self):
        sum = 0
        for val in self.marks:
            sum += val
        print("Hi", self.name, "your avg score is:", sum/3)

s1 = Student("Tony Stark", [99, 98, 97])
s1.get_avg()
s1.hello()
s1.name = "Iron Man"
s1.get_avg()
````

## **Important**

### **Abstraction**

Hiding the implementation details of a class and only showing the essential features to the user.

````python
class Car:
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = False

    def start(self):
        self.clutch = True
        self.acc = True
        print("car started..")

car1 = Car()
car1.start()
````

### **Encapsulation**

Wrapping data and functions into a single unit (object).

````python
class Car:
    def __init__(self):
        self.__acc = False    # private variable
        self.__brk = False
        self.__clutch = False

    def start(self):
        self.__clutch = True
        self.__acc = True
        print("Car started...")

    def stop(self):
        self.__acc = False
        self.__brk = True
        print("Car stopped...")

    # Getter method
    def get_status(self):
        return {
            "Accelerator": self.__acc,
            "brake": self.__brk,
            "Clutch": self.__clutch
        }

# Object
car1 = Car()
car1.start()

print(car1.get_status())
````

### **Inheritance**

The process of inheriting the properties of the parent class into a child class is called inheritance. 

````python
# Parent Class
class Vehicle:
    def start(self):
        print("Vehicle started")

    def stop(self):
        print("Vehicle stopped")

# Class Class
class Car(Vehicle):
    def horn(self):
        print("Car horn: Beep Beep")

# Object
car1 = Car()
car1.start()  # Inherited from Vehicle
car1.horn()   # Own method
````

### **Polymorphism**

Polymorphism allows us to perform the same action in many different ways.

````python
# Parent Class
class Vehicle:
    def start(self):
        print("Vehicle started")

    def stop(self):
        print("Vehicle stopped")

# Class Class
class Car(Vehicle):
    def horn(self):
        print("Car horn: Beep Beep")

# Object
car1 = Car()
car1.start()  # Inherited from Vehicle
car1.horn()   # Own method
````
