### **Private(like) attribute & methods**

**Conceptual Implementation in Python**

Private attributes & methods are meant to be used only within the class and are not accessible from outside the class.


Public Explain

````python
class Account:
    def __init__(self, acc_no, acc_pass):
        self.acc_no = acc_no     # Public Variable (acc_no)
        self.acc_pass = acc_pass  

acc1 = Account("12345", "abc123")

print(acc1.acc_no)
print(acc1.acc_pass)
````

Private Explain

````python
class Account:
    def __init__(self, acc_no, acc_pass):
        self.acc_no = acc_no
        self.__acc_pass = acc_pass  # Private Variable (__acc_pass)

    def reset_pass(self):
        print(self.__acc_pass)    

acc1 = Account("12345", "abc123")

print(acc1.acc_no)
print(acc1.reset_pass())
````

#### **Static Method**

A static method is a method that:

- Belongs to the class

- Does NOT use self

- Cannot access or modify object attributes

- Is created using @staticmethod

It is used for general utility logic related to the class but not dependent on object data.

````python
class Car:
    @staticmethod
    def start():
        print("Car Started..")

    @staticmethod
    def stop():
        print("Car Stopped..")

class ToyotaCar(Car):
    def __init__(self,name):
        self.name=name

car1 = ToyotaCar("Mazda")
car2 = ToyotaCar("Audi")        

car1.start()
````

````python
class Car:
    color = "black"
    @staticmethod
    def start():
        print("Car Started..")

    @staticmethod
    def stop():
        print("Car Stopped..")

class ToyotaCar(Car):
    def __init__(self,name):
        self.name=name

car1 = ToyotaCar("Mazda")
car2 = ToyotaCar("Audi")        

print(car1.color)
car1.start()
````

#### **Instance Method**

An instance method is a method that:

- Belongs to an object (instance) of a class

- Uses self as its first parameter

- Can access and modify the object’s attributes

It works on specific object data.

````python
class Car:
    def start(self):
        print(self.name, "Car Started..")

    def stop(self):
        print(self.name, "Car Stopped..")

class ToyotaCar(Car):
    def __init__(self,name):
        self.name=name

car1 = ToyotaCar("Mazda")
car2 = ToyotaCar("Audi")        

car1.start()
car2.start()
````

### **Super Method**

super() method is used to access methods of the parent class.

````python
class Car:
    def __init__(self,type):
        self.type=type


    @staticmethod
    def start():
        print("Car Started..")

    @staticmethod
    def stop():
        print("Car Stopped.")

class ToyotaCar(Car):
    def __init__(self,name,type):
        super().__init__(type)
        self.name=name
        super().start()

print(car1.type)
car1 = ToyotaCar("prius","electric")
````

### **Property**

We use @property decorator on any method in the class to use the method as a property.

     attribute value -> function # property/attribute

@getter

@setter

````python
class Student:
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math
        self.percentage = str((self.phy + self.chem + self.math) / 3) + "%"

    def calcPercentage(self):
        self.percentage = str((self.phy + self.chem + self.math) / 3) + "%"

stu1 = Student(98, 97, 99)
print(stu1.percentage)

stu1.phy = 86
print(stu1.phy)
stu1.calcPercentage()
print(stu1.percentage)
````

Property Decorator

````python
class Student:
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math

    @property # Property Decorator
    def percentage(self):
        return str((self.phy + self.chem + self.math) / 3) + "%"

stu1 = Student(98, 97, 99)
print(stu1.percentage)

stu1.phy = 86
print(stu1.percentage)
````

##### **Complex Numbers**

1 i + 3j

real   img

````python
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNumber(self):
        print(self.real,"i +", self.img,"j")

    def add(self, num2):
        newReal = self.real + num2.real
        new_Img = self.img + num2.img
        return Complex(newReal, new_Img)

num1 = Complex(1, 3)
num1.showNumber()

num2 = Complex(4, 6)
num2.showNumber()

num3 = num1.add(num2)
num3.showNumber()
````

````python
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNumber(self):
        print(self.real,"i +", self.img,"j")

    def __add__(self, num2):
        newReal = self.real + num2.real
        new_Img = self.img + num2.img
        return Complex(newReal, new_Img)
    
    def __sub__(self, num2):
        newReal = self.real - num2.real
        new_Img = self.img - num2.img
        return Complex(newReal, new_Img)
    
num1 = Complex(1, 3)
num1.showNumber()

num2 = Complex(4, 6)
num2.showNumber()

num3 = num1 - num2
num3.showNumber()
````

````python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius    

c1 = Circle(21)
print(c1.area())
print(c1.perimeter())
````
