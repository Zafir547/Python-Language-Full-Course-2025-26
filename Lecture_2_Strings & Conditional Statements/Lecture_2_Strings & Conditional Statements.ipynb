**Slicing**

Accessing parts of a string

   str[starting_idx : ending_idx] #ending idx is not included

   str = "Apna College"

   str[1 : 4] is "pna"

   str[ : 4] is same as str[0:4]

   str[ : ] is same as str[1 : len(str)]

```python
str = "apna college"
print(str[1:4])
```

```python
str = "apna college"
print(str[0:4])
```

```python
str = "apna college"
print(str[5:12])
```

```python
str = "apna college"
print(str[5:len(str)])
```

```python
str = "apna college"
print(str[:4]) #[0:4]
print(str[5:]) #[5:len(str)]
```

Grade students based on marks

```python
# marks = 74
marks = int(input("enter student marks : "))

if(marks >= 90):
    grade = "A"
elif(marks >= 80 and marks < 90):
    grade = "B"
elif(marks >= 70 and marks < 80):
    grade = "C"
else:
    grade = "D"

print("grade of the student ->", grade)
```

**Nesting**

```python
age = 95

#nesting
if(age >= 18):
    if(age >= 80):
        print("cannot drive")
    else:
        print("can drive")
else:
    print("cannot drive")            
```
