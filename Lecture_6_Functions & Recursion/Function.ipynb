```python
def calc_avg(a,b,c):
    if a == 0:
        total = a+b+c
        avg = total/3
        print("a is zero. Average:",avg)
        return avg
    elif a < 0:
        print("Warning: a is negative. Cannot calculate average.")
        return None
    else:
        total = a + b + c
        avg = total / 3
        print("a is positve. Average:", avg)
        return avg

calc_avg(0, 5, 4)   # a == 0
calc_avg(-1, 5, 4)  # a < 0
calc_avg(2, 5, 4)   # a > 0    
```

```python
def calc_avg(a,b,c):
    # Check if any input is negative
    if a < 0 or b < 0 or c < 0:
        print("Error: Negative values are not allowed.")
        return None
    
    # All inputs are valid (positive or zero)
    total = a + b + c
    avg = round(total/3,2) #Rounded to 2 decimal places

    # Extra message if a is zero
    if a == 0:
        print("Note: a is zero.")

    print(f"Average of {a}, {b}, and {c} is: {avg}")
    return avg    

calc_avg(0, 5, 4)    # a is zero, valid
calc_avg(-1, 5, 4)   # a is negative, invalid
calc_avg(2, 5, 4)    # all positive, valid
```

### **Default Parameters**

Assigning a default value to parameter, which is used when no argument is passed.

```python
def cal_prod(a,b):
    return a * b

cal_prod(2,4)
```

```python
def cal_prod(a=2,b=4): #default value to paramter
    return a * b

cal_prod() # no argument
```

```python
def cal_prod(a,b=2): # default value for b
    return a * b

cal_prod(1)
```

```python
def cal_prod(b,a=2): # default value for a
    return a * b

cal_prod(1)
```

```python
def cal_prod(b,a=2): # default value for a
    return a * b

cal_prod(1)
```

## **Recursion**

When a function calls itself repeatedly.

```python
def show(n):
    print(n)

show(5) #5, 4, 3, 2, 1
```

```python
#recursive function
def show(n):
    if(n == 0):
        return
    print(n)
    show(n-1)

show(5) #5, 4, 3, 2, 1
```

```python
#recursive function
def show(n):
    if(n == 0):
        return
    print(n)
    show(n-1)

show(5) #5, 4, 3, 2, 1
```

```python
#recursive function
def show(n):
    if(n == 1):
        return
    print(n)
    show(n-1)

show(5) #5, 4, 3, 2, 1
```

```python
#recursive function
def show(n):
    if(n == -3):
        return
    print(n)
    show(n-1)

show(5) #5, 4, 3, 2, 1
```

Write a recursive function to calculate the sum of first n natural numbers.

```python
def calc_sum(n):
    if(n == 0):
        return
    print(n)
    calc_sum(n-1)

calc_sum(5)
```

You can add n for this calc_sum(n-1) + n

example: sum(5) -> sum(4) + 5 -> sum(3) + 4 -> sum(2) + 3

```python
def calc_sum(n):
    if(n == 0):
        return 0
    return calc_sum(n-1) + n

sum = calc_sum(5)
print(sum)    
```

Write a recursive function to print all elements in a list.

Hint : use list & index as parameters.

```python
def print_list(list, idx=0):
    if(idx == len(list)):
        return
    print(list[idx])
    print_list(list, idx+1)

fruits = ["mango", "litchi", "apple", "banana"]

print_list(fruits)
```
